Converts code into 1s and 0s for a specific OS and CPU architecture.

### What is an interpreter?

A native program that interprets code at runtime into OS/CPU-specific instructions.

**Compiled** = fast but not portable

**Interpreted** = portable but slow(er)

### Whitespace

- Almost all whitespace is ignored.
- Important whitespace:
  - inside strings
  - around keywords and function/variable names
  - newlines after preprocessor directives (i.e. `#include ...`)
- Most of the formatting you see in a C++ file is for readability
  - Pick a reasonable style and stick to it
  - If you aren't opinionated, follow the style you see in my code

### Braces and Semicolons

In Python, the structure of the code is controlled by whitespace:

```python
def foo(bar):
    if bar > 10:
        while bar > 10:
            bar -= 1
        return bar
    else:
        return bar
```

In C++, you need something else:

- Braces denote blocks of code (similar to python indentation)
- Semicolons denote the end of a statement (similar to a python newline)

```c++
int foo(int bar) { 
    if (bar > 10) { 
        while (bar > 10) { 
            bar--; 
        } 
        return bar; 
    } else { 
        return bar;
    } 
}
```
